```python
# Линейный график 
"""
    Для построения линейного графика используется функция plot()
    plot([x], y, [fmt], *, data=None, **kwargs)
    plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
    plot(y) - график, у которого по оси ординат (y) будут отложены значения из переданного списка,
        по оси х - индексы элементов массива 
        
    arguments:
        x, x2:array - набор данных для оси абсцисс первого, второго и тд графика 
        y, y2:array - набор данных для оси ординат первого, второго и тд графика 
        fmt: str - формат графика '[marker][line][color]'
        **kwargs - свойства класса Line2D - которые предоставляют доступ к большому количеству настроек внешнего вида графика
        
        Line2D - свойства класса: 
        alpha: float - прозрачность 
        color | c: color - цвет 
        fillstyle: {'full', 'left', 'right', 'bottom', 'top', 'none'} - стиль заливки 
        label: object - текстовая метка 
        linestyle | ls: {'-', '--', '-.', ':', '', (offset, on-off-seq), ...} - стиль линии 
        linewidth | lw: float - толщина линии 
        marker: matpLotlib.markers - стиль маркера
        markeredgecolor | mec: color - цвет границы маркера 
        markeredgewidth | mew: float - толщина границы маркера 
        markerfacecolor | mfc: color - цвет заливки маркера 
        markersize | ms: float - размер маркера 
"""
```

```python
# Параметры аргумента fmt - '[marker][line][color]'
"""
    marker: str  - тип маркера 
        одно из [.,ov^<>1234sp*hH+xDd|_]
    line: str - стиль линии 
        одно из [-, --, -., :]
    color: color - цвет графика,
        например, одно из [bgrcmykw]
"""
from matplotlib import pyplot as plt

x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [4, 3, 1, 8, 12]
plt.figure(figsize=(12, 7))
plt.plot(x, y1, 'o-r', alpha=0.7, label='first', lw=5, mec='b', mew=2, ms=10)
plt.plot(x, y2, 'v-.g', label='second', mec='r', lw=2, mew=2, ms=12)
plt.legend()
plt.grid(True)
```

```python
# Заливка области между графиком и осью 
"""
    Для заливки областей используется функция fill_between()
    Сигнатура функции: 
        x: массив длины N - набор данных для оси абсцисс
        y1: массив длины N или скалярное значение - набор данных для оси ординат - первая кривая 
        y2: массив длины N или скалярное значение - набор данных для оси ординат - вторая кривая
        where: массив bool элементов (длины N), по умолчанию None - задает заливаемый цветом регион,
            который определяется координатами x[where]: интервал будет залит между x[i] и x[i+1], если 
            where[i] и where[i+1] равны True 
        step: {'pre', 'post', 'mid'} - определяет шаг, если используется step-функция для отображения графика 
        **kwargs - свойства класса Polygon 
"""

import numpy as np 
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)

plt.plot(x, y, c='r')
plt.fill_between(x, y)
```

```python
# Интервал отображения маркера 
"""
    markevery:
        None - отображаться будет каждая точка 
        N - отображаться будет каждая N-я точка 
        (start, N) - отображается каждая N-я точка, начиная с точки start 
        slice(start, end, N) - отображается каждая N-я точка в интервале от start до end 
        [i, j, m, n] - будут отображены только точки i, j, m, n
"""
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)
m_ev_case = [
    None, 
    10, 
    (100, 30),
    slice(100, 400, 15),
    [0, 100, 200, 300],
    [10, 50, 100]
]

fig, ax = plt.subplots(2, 3, figsize=(10, 7))
axs = [ax[i, j] for i in range(2) for j in range(3)]
for i, case in enumerate(m_ev_case):
    axs[i].set_title(str(case))
    axs[i].plot(x, y, 'o', ls='-', ms=7, markevery=case)
```

```python
# Обрезка графика - для отобрадения части использовать предварительное маскирование данных 
# с помощью функции masked_where() из пакета numpy 
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)
y_masked = np.ma.masked_where(y < -0.5, y)
plt.ylim(-1, 1)
plt.plot(x, y_masked, linewidth=3)
```

```python
# Ступенчатый график 
"""
    step(x, y, [fmt], *, data=None, where='pre', **kwargs)
        x: массив - набор данных для оси абсцисс 
        y: массив - набор данных для оси ординат 
        fmt: str - формат линии
        data: индексируемый объект 
        where: {'pre', 'post', 'mid'} - определяет место, где будет установлен шаг 
            'pre': значение y ставится от значения x, т.е значение y[i] определяется для интервала (x[i-1]; x[i])
            'post': значение y ставится справа от значения x, т.е значение y[i] определяется для интервала (x[i], x[i+1])
            'mid': значение y ставится в середине интервала 
"""

x = np.arange(0, 7)
y = x 

where_set = ['pre', 'post', 'mid']
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

for i, ax in enumerate(axs):
    ax.step(x, y, 'g-o', where=where_set[i])
    ax.grid()
```

```python
"""
    pie()
        x: массив - массив с размерами долей 
        explode: массив - если параметр не None, то часть долей, которые перечислены в передаваемом значении, 
            будут вынесены из диаграммы на заданное расстояние 
        labels: list - текстовые метки долей 
        colors: массив - цвета долей 
        autopct: str - формат текстовой метки внутри доли, 
            текст - численное значение показателя, связанного с конкретной долей 
        pctdistance: float - расстояние между центром каждой доли и началом текстовой метки, которая определяется
            параметром autopct 
        shadow: bool - отображение тени для диаграммы 
        labeldistance: float - расстояние на котором будут отображены текстовые метки долей
        startangle: float - угол, на который нужно повернуть диаграмму против часовой стрелки относительно x 
        radius: float - величина радиуса диаграммы 
        counterclock: bool - определяет направление вращения: по часовой или против часовой стрелки
        wedgeprops: dict - словарь параметров, определяющих внешний вид долей (matplotlib.patches.Wedge)
        textprops: dict - словарь параметров, определяющих внешний вид текстовых меток (matplotlib.text.Text)
        center: list значений float - центр диаграммы 
        frame: bool - рамка внутри диаграммы 
        rotatelabels: bool - если параметр True, то текстовые метки будут повернуты на заданный угол
"""
vals = [24, 17, 53, 21, 35]
labels = ['Ford', 'Toyota', 'BMW', 'Audi', 'Jaguar']
explode = (0.1, 0, 0.15, 0, 0)
fig, ax = plt.subplots()
ax.pie(
    vals, labels=labels, autopct='%1.1f%%', shadow=True,
    explode=explode, wedgeprops={'lw': 1, 'ls': '--', 'edgecolor': 'k'},
    rotatelabels=True
)
ax.axis('equal')
```

```python
# Вложенные круговые диаграммы 
"""
    Состоит из двух компонент: 
        внутренняя - детальное представление информации 
        внешняя - суммарное представление информации по заданным областям 
"""

fig, ax = plt.subplots()
offset = 0.4 
data = np.array([[5, 10, 7], [8, 15, 5], [11, 9, 7]])
cmap = plt.get_cmap('tab20b')
b_colors = cmap(np.array([0, 8, 12]))
sm_colors = cmap(np.array([1, 2, 3, 9, 10, 11, 13, 14, 15]))
ax.pie(
    data.sum(axis=1), 
    radius=1, 
    colors=b_colors,
    wedgeprops=dict(width=offset, edgecolor='w')
)
ax.pie(
    data.flatten(),
    radius=1-offset,
    colors=sm_colors,
    wedgeprops=dict(width=offset, edgecolor='w')
)
```

```python
"""
    imshow()
    X: массив или PIL-изображение 
        (M, N) - двумерный массив со скалярными данными 
        (M, N, 3) - массив с RGB-значениями (0-1 float, 0-255 int)
        (M, N, 4) - массив с RGBA-значениями (0-1 float, 0-255 int)
    cmap: str | Colormap - цветовая карта изображения 
    norm: Normalize - произведение скалярных данныъ к диапазону [0, 1] перед использованием cmap.
        Параметр игнорируется для RGB(A) данных 
    aspect: {'equal', 'auto'} | float 
        equal - соотношение сторон равное 1 
        auto - соотношение не изменяется 
    interpolation: str - алгоритм интерполяции - их очень много, например none, nearest, bilinear, bicubic итд 
    alpha: скалярное значение - прозрачность [0, 1] - игнорируется для RGBA-значения 
    vmin, vmax: скалярное значение - определяют диапазон данных, который будет покрыт цветовой картой, 
        по умолчанию покрывается все 
    origin: {'upper', 'lower'} - расположение начал координат
        upper - верхний левый угол 
        lower - нижний левый угол
    extent: (left, right, bottom, top) - изменение размеров изображения вдоль осей x,y
    filterrad: float>0 - значение по умолчанию 4.0 
        filter radius для фильтров, которые его используют: 'sinc', 'lanczos' или 'blackman'
"""
fig, axs = plt.subplots(1, 2, figsize=(10, 3), constrained_layout=True)

p1 = axs[0].imshow(
    data, cmap='winter', aspect='equal',
    vmin=-1, vmax=1, origin='lower'
)
fig.colorbar(p1, ax=axs[0])

p2 = axs[1].imshow(
    data, cmap='plasma', aspect='equal',
    interpolation='gaussian', origin='lower',
    extent=(0, 30, 0, 30)
)
fig.colorbar(p2, ax=axs[1])
```
