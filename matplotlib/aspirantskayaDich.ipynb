```python
from matplotlib import pyplot as plt
import numpy as np 

data_1 = [88, 73, 55, 42]
data_2 = [71, 68, 44, 24]
labels = ["AKP30", "AKP50", "AKP70", "AKP90"]
x = np.arange(4)

fig, ax = plt.subplots()

width = 0.25

rects1 = ax.bar(x - width/2, data_1, width, label="\"Биологически наивные\"\nn=33", color="blue")
rects2 = ax.bar(x + width/2, data_2, width, label="\"Биологически ненаивные\"\nn=34", color="lightblue")

ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_yticks(range(0, 110, 10))

ax.legend(bbox_to_anchor=(0,-0.3), loc="lower left",  ncol= 2 )

def add_caption_values(bar):
    for p in bar:
        height = p.get_height()
        ax.annotate(
            '{}'.format(height),
            xy=(p.get_x() + p.get_width() / 2, height),
            xytext=(0, 5),
            textcoords="offset points",
            ha='center', 
            va='bottom'
        )
        
add_caption_values(rects1)
add_caption_values(rects2)

probabilities = [0.14, 0.85, 0.54, 0.17]
for i in range(len(rects1)):
    height = rects1[i].get_height() + 8
    x = rects1[i].get_x() + 0.15
    plt.text(x, height, f"p={probabilities[i]}")
```

```python
from matplotlib import pyplot as plt
import numpy as np 

# Вот тут пишешь список своих наблюдений 
data_1 = [88, 73, 55, 42]
data_2 = [71, 68, 44, 24]

# Вот тут пишешь как называется отдельная группа в графике
labels = ["AKP30", "AKP50", "AKP70", "AKP90"]
x = np.arange(len(labels))

fig, ax = plt.subplots()

width = 0.25

# здесь в label подписываешь легенду по каждому из наблюдений, можно цвет выбрать или задать нужный в rgb
rects1 = ax.bar(x - width/2, data_1, width, label="\"Биологически наивные\"\nn=33", color="blue")
rects2 = ax.bar(x + width/2, data_2, width, label="\"Биологически ненаивные\"\nn=34", color="lightblue")

# по оси x задаем количество наблюдений
ax.set_xticks(x)
# и подписываем их
ax.set_xticklabels(labels)
# по оси y говорим промаркируй от 0 до 100
ax.set_yticks(range(0, 110, 10))

# Размещаем легенду под графиком
ax.legend(bbox_to_anchor=(0,-0.3), loc="lower left",  ncol= 2 )

# Вот это вспомогательная штука, чтобы показать значения каждого столбика
def add_caption_values(bar):
    for p in bar:
        height = p.get_height()
        ax.annotate(
            '{}'.format(height),
            xy=(p.get_x() + p.get_width() / 2, height),
            xytext=(0, 5),
            textcoords="offset points",
            ha='center', 
            va='bottom'
        )
        
# Её нужно вызвать для всех наблюдений, чтобы циферки появились        
add_caption_values(rects1)
add_caption_values(rects2)

# Вот здесь пишем вероятности
probabilities = [0.14, 0.85, 0.54, 0.17]

# Вот тут я не разобрался как можно прям ровнехонько по центру турнуть общую группу, 
# можешь изменять пока норм не будет выглядеть
x_offset = 8
y_offset = 0.15

# Подписываем вероятности
for i in range(len(rects1)):
    height = rects1[i].get_height() + x_offset
    x = rects1[i].get_x() + y_offset
    plt.text(x, height, f"p={probabilities[i]}")
```
