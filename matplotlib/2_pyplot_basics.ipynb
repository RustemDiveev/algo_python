```python
# Практически все графики можно построить при помощи модуля pyplot 

import matplotlib.pyplot as plt 
%matplotlib inline 
# Построение графика 
plt.plot()
```

```python
# Если передать список - то значения будут отложены по оси y, а индексы элементов по оси x 
plt.plot([1,7,3,5,11,1])
```

```python
# Чтобы задать значения по обоим осям, в plot() необходимо передать два списка
plt.plot([1, 5, 10, 15, 20], [1, 7, 3, 5, 11])
```

```python
# Наименование осей - функции xlabel(), ylabel() - в дальнейшем, на примере xlabel()
"""
xlabel(
    :str - название оси,
    labelpad:int=None, - расстояние между областью графика и меткой (что такое метка?),
    далее - параметры конструктора класса matplotlib.text.Text
    fontsize|size:int или значение из списка 
        {'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'} - размер шрифта,
    fontstyle: значение из списка: {'normal', 'italic', 'oblique'} - стиль шрифта,
    fontweight: число в диапазоне от 0 до 1000 или значение из списка 
        {'ultralight', 'light', 'normal', 'regular', 'book', 'medium', 'roman', 'semibold',
        'demibold', 'demi', 'bold', 'heavy', 'extra bold', 'black'} - толщина шрифта,
    color - цвет шрифта (будет далее)
)
"""
plt.xlabel('Day', labelpad=20, fontsize=20, color='blue')
plt.plot([1, 5, 10, 15, 20], [1, 7, 3, 5, 11])
```

```python
# Заголовок графика
"""
title(
    :str - заголовок,
    loc - {'center', 'left', 'right'} - выравнивание заголовка,
    параметры конструктора класса matplotlib.text.Text
)
"""
plt.title('Chart price', fontsize=17, loc='left')
plt.plot()
```

```python
# Текстовое примечание 
plt.text(1, 1, 'type: Steel')
# первые два аргумента - координаты позиции надписи, третий - текст надписи 
plt.plot([1,2,3,4,5])
```

```python
# Легенда 

x = [1,5,10,15,20]
y = [1,7,3,5,11]
plt.plot(x, y, label='steel price')
plt.title('Chart price', fontsize=15)
plt.xlabel('Day', fontsize=12, color='blue')
plt.ylabel('Price', fontsize=12, color='blue')
plt.legend()
plt.grid(True)
plt.text(15, 4, 'grow up!')
```

```python
# Параметры отображения графика можно задать как в самой функции plot(), так и при помощи setp()
plt.plot(x, y, color="red")
# Вот это не работает
# plt.setp(color='red', linewidth=1) 
```

```python
# Стиль линии графика - параметр linestyle 
"""
    '-' или 'solid'        - непрерывная линия 
    '--' или 'dashed'      - штриховая линия 
    '-.' или 'dashdot'     - штрихпунктирная линия 
    ':' или 'dotted'       - пунктирная линия 
    'None' или '' или ' '  - не отображать линию 
"""

x = [1, 5, 10, 15, 20]
y = [1, 7, 3, 5, 11]
plt.plot(x, y, '--')
```

```python
# другой вариант, вот здесь setp работает, тк нужно инициализировать экземпляр pyplot и передать его первым параметром 
line = plt.plot(x, y)
plt.setp(line, linestyle='--')
```

```python
# несколько графиков на одном поле 
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [i*1.2 + 1 for i in y1]
y3 = [i*1.2 + 1 for i in y2]
y4 = [i*1.2 + 1 for i in y3]
plt.plot(x, y1, '-', x, y2, '--', x, y3, '-.', x, y4, ':')
```

```python
# тот же самый результат - но вызов plot по отдельности
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [i*1.2 + 1 for i in y1]
y3 = [i*1.2 + 1 for i in y2]
y4 = [i*1.2 + 1 for i in y3]
plt.plot(x, y1, '-')
plt.plot(x, y2, '--')
plt.plot(x, y3, '-.')
plt.plot(x, y4, ':')
```

```python
# цвет линии
"""
    параметр color или с
    значение может быть в одном из следующих форматов:
        RGB или RGBA кортеж значений с плавающей точкой в диапазоне [0, 1], например, (0.1, 0.2, 0.3)
        RGB или RGBA кортеж в hex формате, например, '#0a0a0a'
        строковое представление числа с плавающей точкой в диапазоне [0, 1], например, '0.7'
        символ из набора: {'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'}
        имя цвета из палитры x11/css4 
        цвет из палитры xkcd, должен начинаться с префикса 'xkcd:'
        цвет из набора Tableau Color (палитрп T10), должен начинаться с префикса 'tab:'
"""
# если цвет задается при помощи символа (буквы цвета), то его можно совместить со стилем линии 
x = [1, 5, 10, 15, 20]
y = [1, 7, 3, 5, 11]
plt.plot(x, y, '--r')
```

```python
# тип графика 
# plot() позволяет задать тип графика - линейный или точечный 
# ro - красная о 
plt.plot(x, y, 'ro')
```

```python
# bx - синий x 
plt.plot(x, y, 'bx')
```

```python
# размещение графиков отдельно друг от друга 
"""
    Есть следующие подходы для размещения нескольких графиков на разных полях 
    1) subplot() - указание места для размещения поля с графиком 
    2) subplots() - предварительное задания сетки, в которую будут укладываться поля 
    3) GridSpec - для более гибкого задания геометрии размещения полей с графиками в сетке
"""
```

```python
""" 
    1)
    простейший способ представить графики на отдельных полях использовать subplot 
    subplot(nrows, ncols, index)
        nrows - количество строк 
        ncols - количество столбцов 
        index - местоположение элемента 
    или 
    subplot(pos)
        pos - позиция в виде трехзначного числа, содержащего информацию о количестве строк, столбцов и индексе 
"""
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [i*1.2 + 1 for i in y1]
y3 = [i*1.2 + 1 for i in y2]
y4 = [i*1.2 + 1 for i in y3]

# Настройка размеров подложки - кортеж, передающийся в figsize - высота и ширина подложки 
plt.figure(figsize=(12, 7))

# вывод графиков 
plt.subplot(2, 2, 1)
plt.plot(x, y1, '-')
plt.subplot(2, 2, 2)
plt.plot(x, y2, '--')
plt.subplot(2, 2, 3)
plt.plot(x, y3, '-.')
plt.subplot(2, 2, 4)
plt.plot(x, y4, ':')
```

```python
# второй вариант использования subplot 
plt.figure(figsize=(12, 7))

plt.subplot(221)
plt.plot(x, y1, '-')
plt.subplot(222)
plt.plot(x, y2, '--')
plt.subplot(223)
plt.plot(x, y3, '-.')
plt.subplot(224)
plt.plot(x, y4, ':')
```

```python
"""
    2) 
    subplots() - первые два параметра - количество строк и столбцов сетки 
    возвращает два объекта: 
        1) Figure - подложка, 
        2) Axes - массив объектов для настройки внешнего вида отображаемых компонентов 
"""
fig, axs = plt.subplots(2, 2, figsize=(12, 7))
axs[0, 0].plot(x, y1, '-')
axs[0, 1].plot(x, y2, '--')
axs[1, 0].plot(x, y3, '-.')
axs[1, 1].plot(x, y1, ':')
```
