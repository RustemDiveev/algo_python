```python
import numpy as np
import matplotlib.pyplot as plt

# Магическая команда - хз что это пока
%matplotlib inline
```

```python
# Не знаю numpy вообще
# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
# Эта штука генерирует массив от 0 до 10 из 50 элементов с одинаковой разницей между соседними элементами
x = np.linspace(0,10,50)
y = x 

# Построение графика 
plt.title('Линейная зависимость y = x') # заголовок 
plt.xlabel('x') # ось абсцисс 
plt.ylabel('y') # ось ординат 
plt.grid() # включение отображения сетки 
plt.plot(x, y) # построение графика
```

```python
# изменение типа линии и её цвет 
# r - красный, -- - пунктирная линия
plt.plot(x, y, 'r--')
```

```python
# несколько графиков на одном поле 

# линейная зависимость 
x = np.linspace(0, 10, 50)
y1 = x

# квадратичная зависимость 
y2 = [i**2 for i in x]

# построение графика 
plt.title('Зависимости: y1 = x, y2 = x^2') # заголовок 
plt.xlabel('x') # ось абсцисс 
plt.ylabel('y1, y2') # ось ординат 
plt.grid() # включение отображения сетки 
plt.plot(x, y1, x, y2) # построение графика
```

```python
# представление графиков на разных полях 

# линейная зависимость 
x = np.linspace(0, 10, 50)
y1 = x

# квадратичная зависимость 
y2 = [i**2 for i in x]

# построение графиков 
# figure() - задание глобальных параметров отображения графиков
# передаем кортеж, определяющий размер общего поля
plt.figure(figsize=(9, 9))

# subplot() - задание местоположение поля с графиком
# количество строк, количество столбцов, индекс
plt.subplot(2,1,1)
plt.plot(x, y1) # построение графика 
plt.title('Зависимости: y1 = x, y2 = x^2') # заголовок 
plt.ylabel('y1', fontsize=14) # ось ординат 
plt.grid(True) # включение отображения сетки 

plt.subplot(2,1,2)
plt.plot(x, y2) # построение графика 
plt.xlabel('x', fontsize=14) # ось абсцисс 
plt.ylabel('y2', fontsize=14) # ось ординат 
plt.grid(True) # включение отображения сетки 

```

```python
# категориальные данные (т.е не числовые)
fruits = ['apple', 'peach', 'orange', 'banana', 'melon']
counts = [34, 25, 43, 31, 17]
# Вывод диаграммы
plt.bar(fruits, counts)
plt.title('Fruits!')
plt.xlabel('Fruit')
plt.ylabel('Count')
```

```python
# много что непонятно, но потом разберемся

import matplotlib.pyplot as plt 
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
import numpy as np 

x = np.linspace(0, 10, 10)
y1 = 4 * x
y2 = [i**2 for i in x]

# местоположение поля с графиком
fig, ax = plt.subplots(figsize=(8,6))

# заголовок
ax.set_title('Графики зависимостей: y1=4*x, y2=x^2', fontsize=16)
# заголовки осей
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y1,y2', fontsize=14)
# сетка - крупная и малая
ax.grid(which='major', linewidth=1.2)
ax.grid(which='minor', linestyle='--', color='gray', linewidth=0.5)
# точечный график
ax.scatter(x, y1, c='red', label='y1 = 4*x')
# график - линия
ax.plot(x, y2, label='y2 = x^2')
# легенда
ax.legend()

# вот это не понял что такое
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
# разметка на осях
ax.tick_params(which='major', length=10, width=2)
ax.tick_params(which='minor', length=5, width=1)

plt.show()
```
